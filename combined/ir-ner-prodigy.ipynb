```python
import spacy
c07k=open('./C07K.txt').read().strip()
patent_lines = c07k.split('\n')

nlp_ner = spacy.load("./NER_model/model-best")

colors = {"TECH": "#7DF6D9", "MEDICALCONDITION":"#FFFFFF"}
options = {"colors": colors} 

for line in patent_lines[10000:10010]:
    doc = nlp_ner(line)
    spacy.displacy.render(doc, style="ent", options= options, jupyter=True)
```
